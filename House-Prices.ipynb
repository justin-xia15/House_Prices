## Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data

```python
df = pd.read_csv('USA Housing Dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
(df['price'] == 0).sum()
```

```python
df = df[df['price'] > 0]
```

## EDA

```python
df.describe().transpose()
```

```python
# distribution of house prices
plt.figure(figsize=(12,5))
sns.histplot(data=df, x='price', bins=100)
plt.ticklabel_format(style='plain')
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='price', y='condition', orient='h', width=0.8)
```

```python
plt.figure(figsize=(5,3))
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm')
```

```python
df.corr(numeric_only=True)['price'].sort_values(ascending=False)
```

```python
# sqft_living is the feature most correlated with price
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='sqft_living', y='price', hue='condition', palette='bright')
plt.ticklabel_format(style='plain')
plt.legend(title='Condition', bbox_to_anchor=(1.05,0.8), loc=2);
```

```python
# outliers
upper_boundary = 553062.877289 + 3 * 583686.452244
upper_boundary
```

```python
# remove outliers
df = df[df['price'] < upper_boundary]
```

### House Price Distribution After Removing Outliers

```python
df.describe().transpose()
```

```python
# right skewed normal distribution
plt.figure(figsize=(12,5))
sns.histplot(data=df, x='price', bins=25, kde=True)
plt.ticklabel_format(style='plain')
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df, x='price', y='condition', orient='h', width=0.8)
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='sqft_living', y='price', hue='condition', palette='bright')
plt.ticklabel_format(style='plain')
plt.legend(title='Condition', bbox_to_anchor=(1.05,0.8), loc=2);
```
